```python
import pymysql

for i in range(1,200):
    print(i)

# insert into recipeScore values(email, recipeName, score)
classification = ['육류', '채소류', '해산물', '유제품', '곡류']
method = ['볶음', '끓이기', '부침', '조림', '무침', '비빔', '찜', '절임', '튀김', '삶기', '굽기', '데치기', '회', '기타']

conn = pymysql.connect(host='13.125.102.154', user='hello', password='Csedbadmin!1', db='pytest')
curs = conn.cursor()

sql = "select recipeName, difficulty, Classification, method from recipe"
curs.execute(sql)
recipes = curs.fetchall()

sql = "select recipeName, foodName from ingredients"
curs.execute(sql)
ings = curs.fetchall()

path = "C:/Users/cjfwo/jupyter-졸프/score/"
name = input("시작할 이름:")
scores = []
for recipe in recipes:
    scores.append([recipe[0], 0])
    
while(1):
    print("\n이름: ", name)
    for score in scores:
        score[1] = 0

    while(1):
        print("   0. 나오기, 1. 육류, 2. 채소류, 3. 해산물, 4. 유제품, 5. 곡류")
        answer = input("좋아하는 종류는? ")
        if(answer == "0"):
            break;
        else:
            a = int(answer)
        for i in range(0, 100):
            if(recipes[i][2] == classification[a-1]):
                scores[i][1] = scores[i][1] + 1
                
    while(1):
        print("   0. 나오기, 1. 볶음, 2. 끓이기, 3. 부침, 4. 조림, 5. 무침, 6. 비빔, 7. 찜,\n   8. 절임, 9. 튀김, 10. 삶기, 11. 굽기, 12. 데치기, 13. 회, 14. 기타")
        answer = input("좋아하는 조리법은? ")
        if(answer == "0"):
            break;
        else:
            a = int(answer)
        for i in range(0, 100):
            if(recipes[i][3] == method[a-1]):
                scores[i][1] = scores[i][1] + 1
                
    while(1):
        print("   0. 나오기")
        answer = input("재료 이름은? ")
        if(answer == "0"):
            break;
            
        print("   0. 나오기, 1. 선호(+1), 2. 비선호(-1), 3.제외")
        answer2 = input("재료 선호도는? ")
        if(answer2 == "0"):
            break;
        for ing in ings:
            if(ing[1] == answer):
                for score in scores:
                    if(score[0] == ing[0]):
                        if(answer2 == "1"):
                            score[1] = score[1] + 1
                        elif(answer2 == "2"):
                            if(score[1] == 0):
                                continue
                            score[1] = score[1] - 1
                        elif(answer2 == "3"):
                            score[1] = -100
                
    while(1):
        print("   0. 나오기, 1. easy + 1, 2. easy, normal + 1, 3. normal + 1, 4. hard - 1, 5. normal, hard - 1, 6. hard 제외")
        answer = input("선호 난이도는? ")
        if(answer == "0"):
            break;
        elif(answer == "1"):
            for i in range(0, 100):
                if(recipes[i][1] == "easy"):
                    scores[i][1] = scores[i][1] + 1
        elif(answer == "2"):
            for i in range(0, 100):
                if(recipes[i][1] != "hard"):
                    scores[i][1] = scores[i][1] + 1
        elif(answer == "3"):
            for i in range(0, 100):
                if(recipes[i][1] == "normal"):
                    scores[i][1] = scores[i][1] + 1
        elif(answer == "4"):
            for i in range(0, 100):
                if(recipes[i][1] == "hard"):
                    if(score[1] == 0):
                        continue
                    scores[i][1] = scores[i][1] - 1
        elif(answer == "5"):
            for i in range(0, 100):
                if(recipes[i][1] != "easy"):
                    if(score[1] == 0):
                        continue
                    scores[i][1] = scores[i][1] - 1
        elif(answer == "6"):
            for i in range(0, 100):
                if(recipes[i][1] == "hard"):
                    scores[i][1] = -100                                        

    filename = path + name + '.txt'
    with open(filename, "w") as f:
        for score in scores:
            if (score[1] < -1):
                score[1] = -1
            elif (score[1] == 0):
                score[1] = 0
            elif (score[1] > 5):
                score[1] = 5
                
            if (score[1] != 0):
                f.write("insert into recipeScore values(\'" + name + "\', \'" + score[0] + "\', \'" + str(score[1]) + "\');\n")
            
    name = str(int(name) + 1)
```
